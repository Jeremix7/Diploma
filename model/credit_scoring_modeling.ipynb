## LendingClub

LendingClub is a financial services company headquartered in San Francisco, California. It was the first peer-to-peer lender to register its offerings as securities with the Securities and Exchange Commission (SEC) and offer secondary market credit trading. During its heyday, LendingClub was the world's largest peer-to-peer lending platform. The company reported that up to December 31, 2015, loans in the amount of $15.98 billion were issued through its platform.

## Business Understanding

When a company receives a loan application, it must decide whether to approve the loan based on the applicant's profile. The bank's decision is related to two types of risks:
- If the applicant is likely to repay the loan, the refusal to approve the loan will result in loss of profit for the company;
- If the applicant is likely to fail to repay the loan, i.e. he/she is likely to default, then loan approval may result in financial losses for the company.

The data provided contains information about past loan applicants and whether they had "defaults" or not. The goal is to identify patterns indicating the likelihood of a person defaulting, which can be used to commit actions such as refusing to issue a loan, reducing the loan amount, providing loans (to risky applicants) at a higher interest rate, etc.

When a person applies for a loan, the company can make two types of decisions:
- Approve the loan: if the company approves the loan, there are several possible scenarios for the borrower, described below:
- Fully Paid: the borrower has fully repaid the loan (principal amount and interest rate)
- Current: the borrower is in the process of paying off part of the loan, i.e. the loan term has not yet been completed. These candidates are not marked as "overdue".
    - Charged Off: The borrower has not paid the installments on time for a long period of time, i.e. He/she has been overdue on the loan.
    - In Grace Period: the borrower has a certain time during which he can not make payments on the loan without penalties or negative consequences for the credit history.
    - Default: the borrower has not fulfilled its obligations under the loan agreement, that is, has not made payments on time.
    - Late (16-30 days): the borrower is overdue for a period of 16 to 30 days.
    - Late (31-120 days): the borrower is overdue for a period of 31 to 120 days.
    - Does not meet the credit policy. Status:Fully Paid: The loan was fully repaid, but the borrower did not comply with the lender's credit policy at some point (for example, due to a low credit rating or insufficient income).
    - Does not meet the credit policy. Status:Charged Off: The lender recognized the debt as uncollectible and wrote it off (charged off) because the borrower had not fulfilled his obligations on payments. This usually happens after a long delay period (usually more than 120 days). The debt can be transferred to collectors.
- To refuse a loan: The company refuses the loan (because the candidate does not meet their requirements, etc.). Since the loan was refused, there is no transaction history of these candidates with the company, and therefore this data is not available in the company (and therefore is not considered in this dataset).

Delete variables with more than 80% missing values.

```python
threshold = 0.2

total_rows  = df.count()

missing_data = df.select(
    [(F.count(F.when(F.col(c).isNull(), c)) / total_rows).alias(c) for c in df.columns]
).first()

columns_to_drop = [c for c in df.columns if missing_data[c] > threshold]

df = df.drop(*columns_to_drop)
```

Let's define bad and good loans and add additional columns

```python
bad_loan = ["Charged Off",
            "Default",
            "Does not meet the credit policy. Status:Charged Off",
            "In Grace Period",
            "Late (16-30 days)",
            "Late (31-120 days)"]

df = df.withColumn('loan_describe_int', F.when(F.col('loan_status').isin(bad_loan), 0)\
                                         .otherwise(1))

df = df.withColumn('loan_describe', F.when(F.col('loan_describe_int') == 0, 'bad loan')\
                                     .otherwise('good loan'))

df = df.withColumn('emp_length_int', F.when(F.col('emp_length') == '10+ years', 10)\
                                .when(F.col('emp_length') == '9 years', 9)\
                                .when(F.col('emp_length') == '8 years', 8)\
                                .when(F.col('emp_length') == '7 years', 7)\
                                .when(F.col('emp_length') == '6 years', 6)\
                                .when(F.col('emp_length') == '5 years', 5)\
                                .when(F.col('emp_length') == '4 years', 4)\
                                .when(F.col('emp_length') == '3 years', 3)\
                                .when(F.col('emp_length') == '2 years', 2)\
                                .when(F.col('emp_length') == '1 year', 1)\
                                .when(F.col('emp_length') == '< 1 year', 0.5)\
                                .otherwise(None))

west = ['CA', 'OR', 'UT', 'WA', 'CO', 'NV', 'AK', 'MT', 'HI', 'WY', 'ID']
south_west = ['AZ', 'TX', 'NM', 'OK']
south_east = ['GA', 'NC', 'VA', 'FL', 'KY', 'SC', 'LA', 'AL', 'WV', 'DC', 'AR', 'DE', 'MS', 'TN']
mid_west = ['IL', 'MO', 'MN', 'OH', 'WI', 'KS', 'MI', 'SD', 'IA', 'NE', 'IN', 'ND']
north_east = ['CT', 'NY', 'PA', 'NJ', 'RI', 'MA', 'MD', 'VT', 'NH', 'ME']

df = df.withColumn('region', F.when(F.col('addr_state').isin(west), 'west')\
                                .when(F.col('addr_state').isin(south_west), 'south west')\
                                .when(F.col('addr_state').isin(south_east), 'south east')\
                                .when(F.col('addr_state').isin(mid_west), 'mid west')\
                                .when(F.col('addr_state').isin(north_east), 'north east')\
                                .otherwise(None))
```

### Binary Encoding

```python
binary_cat_cols = [
    col for col in fe_df.columns 
    if fe_df.schema[col].dataType == StringType() and fe_df.select(col).distinct().count() == 2
]

print(binary_cat_cols)
```

```python
from pyspark.ml.feature import StringIndexer

for col in binary_cat_cols:
    indexer = StringIndexer(inputCol=col, outputCol=f"{col}_binary", handleInvalid="skip")
    fe_df = indexer.fit(fe_df).transform(fe_df)

fe_df = fe_df.drop(*binary_cat_cols)

fe_df.show(5)
```

```python
fe_df = spark.read.parquet('intermediate_data/fe_df.parquet')
```

```python
multi_cat_cols = [
    col for col, dtype in fe_df.dtypes
    if dtype not in ("float", "int", "double")
    and fe_df.select(col).distinct().count() > 2
]

removed_columns = ['issue_d', 'loan_status', 'addr_state', 'earliest_cr_line', 'last_pymnt_d', 'last_credit_pull_d']
multi_cat_cols = [col for col in multi_cat_cols if col not in removed_columns]
```

```python
for col in multi_cat_cols:
    print(col)
```

```python
multi_cat_cols = ['purpose']

global_mean = fe_df.agg(F.mean("loan_describe_binary")).first()[0]

alpha = 0.2

for col in multi_cat_cols:
    stats = fe_df.groupBy(col).agg(
        F.sum("loan_describe_binary").alias("sum_target"),
        F.count("loan_describe_binary").alias("count_target")
    )

    stats = stats.withColumn(
        f"{col}_encoded",
        (F.col("sum_target") + global_mean * alpha) / (F.col("count_target") + alpha)
    ).select(col, f"{col}_encoded")

    fe_df = fe_df.join(stats, on=col, how="left")

fe_df = fe_df.drop(*[col for col in multi_cat_cols if not col.endswith("_encoded")])

fe_df.show(5)
```

### Train ML model

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
                    .master("local[*]") \
                    .config("spark.executor.memory", "8g") \
                    .config("spark.driver.memory", "4g") \
                    .config("spark.sql.autoBroadcastJoinThreshold", "-1") \
                    .getOrCreate()
```

```python
train_test_data = spark.read.parquet('cleaned_df_before_2018.parquet')
train_test_data.count()
```

```python
train_test_data.printSchema()
```

Let's define the target feature and others.

```python
all_columns = train_test_data.columns
target_column = "loan_describe_int"
feature_columns = [col for col in all_columns if col != target_column]
```

Let's encode categorical features.

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoder
from pyspark.ml import Pipeline

categorical_columns = ["term"]

indexers = [StringIndexer(inputCol=col, outputCol=f"{col}_index", handleInvalid="keep") for col in categorical_columns]

encoders = [OneHotEncoder(inputCol=f"{col}_index", outputCol=f"{col}_encoded") for col in categorical_columns]

pipeline = Pipeline(stages=indexers + encoders)

df_encoded = pipeline.fit(train_test_data).transform(train_test_data)
```

```python
from pyspark.ml.feature import VectorAssembler

numeric_columns = [
    "last_fico_range_high", "last_fico_range_low", "int_rate", "fico_range_low",
    "acc_open_past_24mths", "dti", "num_tl_op_past_12m", "loan_amnt",
    "mort_acc", "avg_cur_bal", "bc_open_to_buy", "num_actv_rev_tl"
]

encoded_columns = [f"{col}_encoded" for col in categorical_columns]
final_feature_columns = numeric_columns + encoded_columns
assembler = VectorAssembler(inputCols=final_feature_columns, outputCol="features")
final_data = assembler.transform(df_encoded).select("features", target_column)


final_data.show(truncate=False, n=5)
```

Let's divide the dataframe into a test and a train.

```python
train_data, test_data = final_data.randomSplit([0.8, 0.2], seed=42)
```

```python
from pyspark.ml.classification import LogisticRegression
from pyspark.ml.evaluation import BinaryClassificationEvaluator

lr = LogisticRegression(featuresCol="features", labelCol=target_column)

lr_model = lr.fit(train_data)
lr_predictions = lr_model.transform(test_data)

evaluator = BinaryClassificationEvaluator(labelCol=target_column, rawPredictionCol="rawPrediction", metricName="areaUnderROC")
lr_auc = evaluator.evaluate(lr_predictions)

print(f"Logistic Regression - AUC-ROC: {lr_auc:.4f}")
```

```python
from pyspark.ml.classification import DecisionTreeClassifier

dt = DecisionTreeClassifier(featuresCol="features", labelCol=target_column)

dt_model = dt.fit(train_data)
dt_predictions = dt_model.transform(test_data)

evaluator = BinaryClassificationEvaluator(labelCol=target_column, rawPredictionCol="rawPrediction", metricName="areaUnderROC")
dt_auc = evaluator.evaluate(dt_predictions)

print(f"Decision Tree - AUC-ROC: {dt_auc:.4f}")
```

```python
from pyspark.ml.classification import RandomForestClassifier

rf = RandomForestClassifier(featuresCol="features", labelCol=target_column)
rf_model = rf.fit(train_data)

rf_predictions = rf_model.transform(test_data)

evaluator = BinaryClassificationEvaluator(labelCol=target_column, rawPredictionCol="rawPrediction", metricName="areaUnderROC")
rf_auc = evaluator.evaluate(rf_predictions)

print(f"Random Forest - AUC-ROC: {rf_auc:.4f}")
```

```python
from pyspark.ml.classification import GBTClassifier

gbt = GBTClassifier(featuresCol="features", labelCol=target_column)

gbt_model = gbt.fit(train_data)
gbt_predictions = gbt_model.transform(test_data)

evaluator = BinaryClassificationEvaluator(labelCol=target_column, rawPredictionCol="rawPrediction", metricName="areaUnderROC")
gbt_auc = evaluator.evaluate(gbt_predictions)

print(f"Gradient-Boosted Trees - AUC-ROC: {gbt_auc:.4f}")
```

```python
results = {
    "Logistic Regression": lr_auc,
    "Decision Tree": dt_auc,
    "Random Forest": rf_auc,
    "Gradient-Boosted Trees": gbt_auc
}

print("\nModel Comparison:")
for model_name, auc in results.items():
    print(f"{model_name}: AUC-ROC = {auc:.4f}")
```

```python
lr = LogisticRegression(featuresCol="features", labelCol=target_column)

lr_model = lr.fit(train_data)

lr_predictions = lr_model.transform(test_data)

evaluator = BinaryClassificationEvaluator(labelCol=target_column, rawPredictionCol="rawPrediction", metricName="areaUnderROC")
lr_auc = evaluator.evaluate(lr_predictions)

print(f"Logistic Regression - AUC-ROC: {lr_auc:.4f}")

model_save_path = "logistic_regression_model4"
lr_model.save(model_save_path)
print(f"Model saved to {model_save_path}")
```

```python
from pyspark.ml.classification import LogisticRegressionModel

model_save_path = "logistic_regression_model4"
loaded_model = LogisticRegressionModel.load(model_save_path)
predictions = loaded_model.transform(test_data)

print("Sample predictions:")
predictions.select("features", "rawPrediction", "probability", "prediction").show(5)

evaluator = BinaryClassificationEvaluator(labelCol=target_column, rawPredictionCol="rawPrediction", metricName="areaUnderROC")
auc = evaluator.evaluate(predictions)

print(f"logistic_regression_model - AUC-ROC: {auc:.4f}")
```
