## Business Understanding

Когда компания получает заявку на получение кредита, она должна принять решение об одобрении кредита на основе профиля заявителя. Решение банка связано с двумя типами рисков:
- Если заявитель, скорее всего, погасит кредит, то отказ в одобрении кредита приведет к потере прибыли для компании;
- Если заявитель, скорее всего, не погасит кредит, т.е. он / она, скорее всего, допустит дефолт, то одобрение кредита может привести к финансовым потерям для компании.

Приведенные данные содержат информацию о прошлых заявителях на получение кредита и о том, были ли у них "дефолты’ или нет. Цель состоит в том, чтобы выявить закономерности, указывающие на вероятность дефолта лица, которые могут быть использованы для совершения таких действий, как отказ в выдаче кредита, уменьшение суммы кредита, предоставление кредитов (рискованным заявителям) по более высокой процентной ставке и т.д.

Когда человек подает заявку на получение кредита, компания может принять два типа решений:
- Oдобрить кредит: если компания одобрит кредит, есть несколько возможных сценариев для заемщика, описанных ниже:
    - Fully Paid: заемщик полностью выплатил кредит (основную сумму и процентную ставку)
    - Current: заемщик находится в процессе выплаты части кредита, т. е. срок действия кредита еще не завершен. Эти кандидаты не помечены как "просроченные".
    - Charged Off: заемщик не выплачивал взносы в установленные сроки в течение длительного периода времени, т. е. Он / она допустил просрочку по кредиту.
    - In Grace Period: заемщик имеет определенное время, в течение которого он может не производить платежи по кредиту без штрафных санкций или негативных последствий для кредитной истории.
    - Default: заемщик не выполнил свои обязательства по кредитному договору, то есть не произвел платежи в установленный срок.
    - Late (16-30 days): заемщик просрочил платеж на срок от 16 до 30 дней.
    - Late (31-120 days): заемщик просрочил платеж на срок от 31 до 120 дней.
    - Does not meet the credit policy. Status:Fully Paid: кредит был полностью погашен, но заемщик не соответствовал кредитной политике кредитора в какой-то момент (например, из-за низкого кредитного рейтинга или недостаточного дохода).
    - Does not meet the credit policy. Status:Charged Off: кредитор признал долг безнадежным и списал его (charged off), поскольку заемщик не выполнил свои обязательства по платежам. Это обычно происходит после длительного периода просрочки (обычно более 120 дней). Долг может быть передан коллекторам.
- Отказать в выдаче кредита: компания отказывает в кредите (потому что кандидат не соответствует их требованиям и т.д.). Поскольку в кредите было отказано, история транзакций этих кандидатов с компанией отсутствует, и поэтому эти данные недоступны в компании (и, следовательно, в этом наборе данных не рассматриваются).

```python
# Определяем порог для отсутствующих значений
threshold = 0.2

# Получаем общее количество строк в DataFrame
total_rows  = df.count()

# Вычисляем долю отсутствующих значений для каждого столбца
missing_data = df.select(
    [(F.count(F.when(F.col(c).isNull(), c)) / total_rows).alias(c) for c in df.columns]
).first()

# Получаем список столбцов, которые нужно удалить
columns_to_drop = [c for c in df.columns if missing_data[c] > threshold]

# Удаляем столбцы
df = df.drop(*columns_to_drop)
```

Let's define bad and good loans and add additional columns

```python
# Список статусов кредита, которые считаются плохими
bad_loan = ["Charged Off",
            "Default",
            "Does not meet the credit policy. Status:Charged Off",
            "In Grace Period",
            "Late (16-30 days)",
            "Late (31-120 days)"]

# Создание новой колонки 'loan_describe_int', где 0 означает плохой кредит, а 1 - хороший
df = df.withColumn('loan_describe_int', F.when(F.col('loan_status').isin(bad_loan), 0)\
                                         .otherwise(1))

# Создание новой колонки 'loan_describe', где 'bad loan' соответствует 0, а 'good loan' - 1
df = df.withColumn('loan_describe', F.when(F.col('loan_describe_int') == 0, 'bad loan')\
                                     .otherwise('good loan'))

# Создание новой колонки 'emp_length_int', которая преобразует текстовое представление длины работы в числовое
df = df.withColumn('emp_length_int', F.when(F.col('emp_length') == '10+ years', 10)\
                                .when(F.col('emp_length') == '9 years', 9)\
                                .when(F.col('emp_length') == '8 years', 8)\
                                .when(F.col('emp_length') == '7 years', 7)\
                                .when(F.col('emp_length') == '6 years', 6)\
                                .when(F.col('emp_length') == '5 years', 5)\
                                .when(F.col('emp_length') == '4 years', 4)\
                                .when(F.col('emp_length') == '3 years', 3)\
                                .when(F.col('emp_length') == '2 years', 2)\
                                .when(F.col('emp_length') == '1 year', 1)\
                                .when(F.col('emp_length') == '< 1 year', 0.5)\
                                .otherwise(None))  # Присваиваем None, если значение не соответствует ни одному из условий

# Определение списков штатов для различных регионов
west = ['CA', 'OR', 'UT', 'WA', 'CO', 'NV', 'AK', 'MT', 'HI', 'WY', 'ID']
south_west = ['AZ', 'TX', 'NM', 'OK']
south_east = ['GA', 'NC', 'VA', 'FL', 'KY', 'SC', 'LA', 'AL', 'WV', 'DC', 'AR', 'DE', 'MS', 'TN']
mid_west = ['IL', 'MO', 'MN', 'OH', 'WI', 'KS', 'MI', 'SD', 'IA', 'NE', 'IN', 'ND']
north_east = ['CT', 'NY', 'PA', 'NJ', 'RI', 'MA', 'MD', 'VT', 'NH', 'ME']

# Создание новой колонки 'region', которая определяет регион на основе штата адреса
df = df.withColumn('region', F.when(F.col('addr_state').isin(west), 'west')\
                                .when(F.col('addr_state').isin(south_west), 'south west')\
                                .when(F.col('addr_state').isin(south_east), 'south east')\
                                .when(F.col('addr_state').isin(mid_west), 'mid west')\
                                .when(F.col('addr_state').isin(north_east), 'north east')\
                                .otherwise(None))  # Присваиваем None, если штат не соответствует ни одному из регионов
```

```python
# Вычисляем количество пропущенных значений в каждой колонке
missing_values = df.select([F.sum(F.col(c).isNull().cast("int")).alias(c) for c in df.columns])

missing_values.show()
```

```python
# Определение оконной функции для медианы по регионам
window_spec = Window.partitionBy("region")

# Замена пропущенных значений в колонке `emp_length_int` на медиану
median_df = df.groupBy("region").agg(F.expr("percentile_approx(emp_length_int, 0.5)").alias("emp_length_median"))
df = df.join(median_df, on="region", how="left").withColumn(
    "emp_length_int",
    F.when(F.col("emp_length_int").isNull(), F.col("emp_length_median")).otherwise(F.col("emp_length_int"))
).drop("emp_length_median")  # Удаляем временную колонку

# Замена пропущенных значений в колонке `last_pymnt_d` на моду
mode_df = (
    df.groupBy("region", "last_pymnt_d")
    .agg(F.count("*").alias("cnt"))
    .withColumn("rank", F.row_number().over(Window.partitionBy("region").orderBy(F.desc("cnt"))))
    .filter(F.col("rank") == 1)
    .select(F.col("region").alias("mode_region"), F.col("last_pymnt_d").alias("last_pymnt_mode"))
)
df = df.join(mode_df, df["region"] == mode_df["mode_region"], how="left").withColumn(
    "last_pymnt_d",
    F.when(F.col("last_pymnt_d").isNull(), F.col("last_pymnt_mode")).otherwise(F.col("last_pymnt_d"))
).drop("last_pymnt_mode", "mode_region")  # Удаляем временные колонки

# Вывод результата
df.show(5)
```

```python
# Получаем колонки с 2 или менее уникальными значениями
dual_cat_cols = []
multi_cat_cols = []

for col in cat_columns:
    unique_count =  df.select(col).distinct().count()
    if unique_count <= 2:
        dual_cat_cols.append(col)
    else:
        multi_cat_cols.append(col)
```

### Binary Encoding

```python
from pyspark.ml.feature import StringIndexer

# Определяем бинарные категориальные колонки
binary_cat_cols = [
    col for col in df.columns 
    if df.select(col).distinct().count() == 2  # Оставляем только колонки с 2 уникальными значениями
]

# Закодируем каждую бинарную категориальную колонку
for col in binary_cat_cols:
    indexer = StringIndexer(inputCol=col, outputCol=f"{col}_binary", handleInvalid="skip")
    df = indexer.fit(df).transform(df)
```

```python
# Удаляем оригинальные бинарные колонки
df = df.drop(*binary_cat_cols)

# Вывод результата
df.show()
```

```python
from pyspark.ml.feature import StringIndexer

# Определяем мультикатегориальные колонки
multi_cat_cols = [
    col for col in df.columns 
    if df.select(col).distinct().count() > 2  # Оставляем только колонки с более чем 2 уникальными значениями
]

# Преобразуем мультикатегориальные колонки в числовые индексы
for col in multi_cat_cols:
    indexer = StringIndexer(inputCol=col, outputCol=f"{col}_index", handleInvalid="skip")
    df = indexer.fit(df).transform(df)

# Удаляем оригинальные колонки, если больше не нужны
df = df.drop(*multi_cat_cols)

# Обновляем список мультикатегориальных колонок
multi_cat_cols = [f"{col}_index" for col in multi_cat_cols]

# Вывод результата
df.show()
```
